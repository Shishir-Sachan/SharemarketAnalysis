```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.6f}'.format
```

```python
Ved = pd.read_csv("Ved.csv", index_col="Date", parse_dates=["Date"])
Ved
```

```python
Ved["log_ret"] = np.log(Ved.Price / Ved.Price.shift()) # daily log returns
```

```python
Ved
```

```python
Ved.describe()
```

```python
mu = Ved.log_ret.mean() # mean log return ---> Reward
mu
```

```python
sigma = Ved.log_ret.std() # standard deviation of log returns --> Risk/Volatility
sigma
```
