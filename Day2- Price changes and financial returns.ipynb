#  Price changes and financial returns

```python
#  More meaningful/useful than prices: price changes
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
# plt.style.use("seaborn")
plt.style.use("seaborn-v0_8")
```

```python
close = pd.read_csv("close.csv", index_col = "Date", parse_dates=["Date"])
close
```

```python
Ved = close["VEDL.NS"].dropna().to_frame().copy()
```

```python
Ved
```

```python
Ved.rename(columns={"VEDL.NS": "Price"}, inplace = True)
```

```python
Ved
```

```python
Ved.shift(periods = 1)
```

```python
Ved["P_lag1"] = Ved.shift(periods = 1)
```

```python
Ved
```

### Absolute Price Changes (Difference)

```python
Ved["P_diff"] = Ved.Price.sub(Ved.P_lag1) #Alernative 1
```

```python
Ved
```

```python
# Ved.P_diff.drop
```

```python
Ved["P_diff2"] = Ved.Price.diff(periods = 1) # Alternative 2
```

```python
Ved
```

```python
Ved.P_diff.equals(Ved.P_diff2)
```

```python
# Absolute Price changes are not meaningful
```

```python
# Relative/Percentage Price Changes(Returns)
```

```python
Ved.Price.div(Ved.P_lag1) - 1 # Alternative 1
```

```python
Ved["Returns"] = Ved.Price.pct_change(periods = 1) # Alternative 2
```

```python
Ved
```

```python
202.0000 / 199.8500 - 1
```

```python
(202.0000 / 199.8500 - 1) * 100
```

```python
# Take Home: Relative Price Changes (Returns) are meaningful and comparable across instruments
```
