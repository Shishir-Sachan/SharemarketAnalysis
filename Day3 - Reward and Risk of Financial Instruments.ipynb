###  Measuring reward and risk of an investment

```python
# General rule in finance/investing: Higher risk must reward higher returns.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8")
```

```python
Ved = pd.read_csv("Ved.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
Ved
```

```python
Ved.Price.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize =13)
plt.show()
```

Reward: Positive Returns 

Risk: Volatility of returns

```python
Ved.describe()
```

```python
mu = Ved.Returns.mean() # arithmetic mean return ---> reward
```

```python
mu
```

```python
sigma = Ved.Returns.std() # standard deviation of returns ---> Risk/Volatility
```

```python
sigma
```

```python
np.sqrt(Ved.Returns.var()) # standard deviation
```
